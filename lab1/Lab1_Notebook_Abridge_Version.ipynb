## Introduction

Measuring the energy of particles is crucial to many physics experiements. This is usually done by looking at the interaction of the particles with a medium that is well understood. A common apparatus used is a scintillator, which can convert gamma radiation to lower-energy photons, which can be easily detected by phototubes. Phototubes can then covert the low energy photons into electrical signals that are easier to process. 

Scintillator : a material that fluoresces when struck by a charged particle or high-energy photon. (Google definitions)

## Background

Gamma rays from the source will hit the scintillator to be coverted to low energy photons. The photons will be collected by the phototube and converted to electric current. The overall effciency for converting gamma ray energy to electric current is on the order of 0.15%. Each gamma ray will therefore produce an electrical pulse of amplitude proportional the to the energy absorbed by the detectors. Below are some processes of gamma ray conversion that results in different energy absorbed.

Here photopeak will refer to amplitude created when all of the energy is absorbed. 

### Photoelectric effect

Photon is completely absorbed. This process will create a spike at the photopeak.

### Compton effect

Photon bounces off and only transfer part of the energy. The photon could eventually leave the scintillator with some energy, so this process will sometimes be less than the photospeak

### Electron-Positron Pair Production

Some energy goes into excitation of electrons, the rest is converted to two gamma rays each with 0.511MeV. Depending on whether the gamma rays is absorbed, the signal could be at the photopeak, 0.511 MeV or 1.022Mev less than the photopeak.

### Backscatter

Photon pass through the detector, but bounce of the enclosure and back into the detector.

### Compton Scattering

Calculating the compton edge. This is when the most amount of energy can be deposited in the detector.

From equation 4.11 in the lab manual, we have 

$$ \frac{1}{E_o} - \frac{1}{E_i} = \frac{1}{mc^2}(1-\cos\theta)$$

Here, the most amount of energy transferred when $E_o$ is small. Then rearranging some algebra

$$ E_o = \frac{1}{ \frac{1}{E_i} + \frac{1}{mc^2}(1-\cos\theta)} = \frac{E_i}{ 1 + \frac{E_i}{mc^2}(1-\cos\theta)}$$

We see that the smallest energy is when $1-\cos\theta = 2$, or when $\theta = 180^\circ$. Then the total transferred energy given by $E_T = E_i - E_o$ can be written as: 

$$E_T = E_i (1 - \frac{1}{ 1 + \frac{2E_i}{mc^2}})$$

## Final answers to questions

* Why and where does Compton's Edge occur?

The hypotheuse is that there is a maximum amount of energy a photon can transfer during the Compton Scattering process. From energy conservation, we see that the energy transferred depends on the geometry of the collision between the photon and the electron. When the photon leaves the collision in the direction it came from that is when the most energy is transferred. Through this experiement we can see that this theorhetical prediction matches with what we observed in the data. 

* Why and where does the backscattering occur?

The hypotheuse is that compton scattering occurs in the enclosure of the detector, the photon then leaves the enclosure and gets back into the detector. Based on the geometry of the detector and the enclosure, the photon would have to leave the enclosure in the opposite direction in order to go back into the detector. This means the photon would leave approximately the energy of Compton's Edge before returning to the detector. From the experiments, we see that the backscattering peak is actually just a little bit above the initial energy subtracting the Compton's Edge energy. This is most likely due to the fact that the photon does not have to go completely in the opposite direction but rather just get close enough. 

* Why and where does the electron positron pair production occur?

The hypotheuse is that only one of the gamma rays produced from the annihiltion of the electron positron pair would be detected. During this annihilation two gamma rays of equal energy would be produced. Based on momentum conservation, the two gamma rays must travel in the opposite direction. Based on energy conservation, the energy of the two gamma rays combined should be equivalent to the energy of the electron positron pair. These two electron and positron have minimal amount of energy, so their combined energy is roughly just the rest energy of the two particles. Therefore the detector will only ever detect one of the two gamma rays at roughly 0.511MeV. From the experiments, this is exactly what we observed, there is a peak at 0.511MeV but none at 1.022MeV. Furthermore, as gamma ray produced is just like any other gamma ray, it is possible for it to undergo Compton Scattering and even form a backscattering peak, which we observe in the experiment.
